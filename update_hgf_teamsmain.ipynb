# Update elocalculator/clean_data/hgf_matches_ordered.csv

```python
# create DB from csv
matches=pd.read_csv(r"clean_data/hgf_matches.csv",skipinitialspace=True,header=0,dtype={'ID':np.int32,'GT1':np.int32,'GT2':np.int32})
matches
```

```python
# Database of individual teams
teams=pd.read_csv(r"clean_data/hgf_teams.csv",skipinitialspace=True,header=0,names=['OrigName','MainName','TeamID'])
teams
```

```python
# Add MainName and TeamID to the matches DB
nm = len(matches)
nt = len(teams)

# add new columns
matches['T1ID']=pd.Series([0]*nm,index=np.arange(nm))
matches['T2ID']=pd.Series([0]*nm,index=np.arange(nm))


for im in range(nm):
    t1 = matches['Team1'].iloc[im]
    t2 = matches['Team2'].iloc[im]
    for it in range(nt):
        altname = teams['OrigName'].iloc[it]
        tid = teams['TeamID'].iloc[it]
        
        if t1==altname:
            matches['T1ID'].iloc[im]=tid
        if t2==altname:
            matches['T2ID'].iloc[im]=tid
            
matches
```

```python
# Remove unncessary columns from the matches DB
shortmatches = matches
del shortmatches['Team1']
del shortmatches['Team2']
del shortmatches['HGF']
del shortmatches['DGF']
shortmatches
```

```python
# save the smaller DataFrame with matches
# WARNING: writes over "clean_data/hgf_matches_ordered.csv"!
shortmatches.to_csv("clean_data/hgf_matches_ordered.csv")
```
